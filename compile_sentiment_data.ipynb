Cleanup for black lives matter data

```
df_dir = file_dir + 'blacklivesmattersentiment/data.csv'
```

```
df_blm_raw = pd.read_csv(df_dir)
```

```
df_blm_raw.head()
```

```
df_blm=df_blm_raw.copy()
```

```
df_blm = df_blm.drop(labels = 'Location', axis = 1)
```

```
df_blm
```

```
#Search for neutral(Sentiment )) and close-to-neutral tweets
for index, row in df_blm.iterrows():
  if row['Sentiment'] >= -0.02 and row['Sentiment'] <=0.02:
    df_blm.at[index, 'Sentiment'] = int(0)
```

```
df_blm
```

```
for index, row in df_blm.iterrows():
  if row['Sentiment'] == 0:
    df_blm = df_blm.drop(index)
```

```
df_blm
```

```
for index, row in df_blm.iterrows():
  if row['Sentiment'] > 0.02:
    df_blm.at[index, 'Sentiment'] = int(1)
```

```
df_blm
```

```
for index, row in df_blm.iterrows():
  if row['Sentiment'] < -0.02:
    df_blm.at[index, 'Sentiment'] = int(0)
```

```
df_blm
```

```
df_blm.columns = ['tweet', 'label']
```

```
df_blm
```

```
#blm cleaning complete 
```
